Завдання 7 Географічний аналіз компенсації Python-розробників

```python
# Відфільтровуємо респондентів, які працюють з Python
python_devs = df[df["worked_with_python"]]
python_devs = python_devs.dropna(subset=["ConvertedCompYearly"])

# Групуємо по країнах та обчислюємо середню і медіанну компенсацію
compensation_stats = (
    python_devs.groupby("Country")["ConvertedCompYearly"]
    .agg(["mean", "median"])
    .reset_index()
)
# Округлюємо значення для зручності
compensation_stats["mean"] = compensation_stats["mean"].round(2)
compensation_stats["median"] = compensation_stats["median"].round(2)

# Виводимо результат
print("Географічний аналіз компенсації Python-розробників:")
print(compensation_stats)
```

Завдання 9 Аналіз популярності Python по віковим категоріям 

```python
# Групуємо по вікових категоріях
age_stat = df.groupby("Age")["worked_with_python"].mean() * 100
age_stat = age_stat.reset_index()
age_stat.columns = ['Age', 'Percent_python']

#Проводимо округлення
age_stat['Percent_python'] = age_stat['Percent_python'].round(1)

# Виводимо результат
print('Відсоток респондентів, які програмують на Python за віковими категоріями:')
age_stat
```

Завдання 10 Аналіз індустрій серед високооплачуваних віддалених працівників

```python
# Розраховуємо 75-й перцентиль компенсації
q75 = df["ConvertedCompYearly"].quantile(0.75)

# Фільтруємо респондентів
high_paid_remote = df[(df["ConvertedCompYearly"] > q75) &  (df["RemoteWork"] == "Remote")]

# Підраховуємо найпопулярніші індустрії
industry_stats = (high_paid_remote["Industry"].value_counts().reset_index())

# Виводимо результат
print("Найрозповсюдженіші  індустрії серед високооплачуваних віддалених працівників:")
industry_stats
```
